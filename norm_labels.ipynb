```python
BASE_DIR = "D:\\_hack\\2021-22\\d-hacks-ai.ru\\758258-Tatarstan\\data_for_yolo\\data"
train_original_labels_dir = os.path.join(BASE_DIR, "labels", "train-original")
train_new_labels_dir = os.path.join(BASE_DIR, "labels", "train")
train_images_dir = os.path.join(BASE_DIR, "images", "train")
```

```python
# car': 0, 'head': 1, 'face': 2, 'human': 3, 'carplate': 4
for file_name in os.listdir(train_images_dir):
    filename_wo_ext = str(file_name).split(".")[0]
    
    new_data = ""
    try:
        label = 0
        label_filename = filename_wo_ext + "_car..txt"
        file_name_orig = os.path.join(train_original_labels_dir, label_filename)
        f = open(file_name_orig, "r")
        data = str(f.read()).split("\n")
        for line in data:
            l = str(line).split(" ")
            if (len(str(line)) > 1):
                l[0] = str(label)
                new_data += " ".join(l)
                new_data += "\n"
        f.close()

        print("exec")
    except:
        pass
        print(file_name_orig)
        print("no file")

    try:
        label = 1
        label_filename = filename_wo_ext + "_head..txt"
        file_name_orig = os.path.join(train_original_labels_dir, label_filename)
        f = open(file_name_orig, "r")
        data = str(f.read()).split("\n")
        for line in data:
            l = str(line).split(" ")
            if (len(str(line)) > 1):
                l[0] = str(label)
                new_data += " ".join(l)
                new_data += "\n"
        f.close()
    except:
        pass

    try:
        label = 2
        label_filename = filename_wo_ext + "_face..txt"
        file_name_orig = os.path.join(train_original_labels_dir, label_filename)
        f = open(file_name_orig, "r")
        data = str(f.read()).split("\n")
        for line in data:
            l = str(line).split(" ")
            if (len(str(line)) > 1):
                l[0] = str(label)
                new_data += " ".join(l)
                new_data += "\n"
        f.close()
    except:
        pass

    try:
        label = 3
        label_filename = filename_wo_ext + "_human..txt"
        file_name_orig = os.path.join(train_original_labels_dir, label_filename)
        f = open(file_name_orig, "r")
        data = str(f.read()).split("\n")
        for line in data:
            l = str(line).split(" ")
            if (len(str(line)) > 1):
                l[0] = str(label)
                new_data += " ".join(l)
                new_data += "\n"
        f.close()
    except:
        pass

    try:
        label = 4
        label_filename = filename_wo_ext + "_carplate..txt"
        file_name_orig = os.path.join(train_original_labels_dir, label_filename)
        f = open(file_name_orig, "r")
        data = str(f.read()).split("\n")
        for line in data:
            l = str(line).split(" ")
            if (len(str(line)) > 1):
                l[0] = str(label)
                new_data += " ".join(l)
                new_data += "\n"
        f.close()
    except:
        pass


    file_name_new = filename_wo_ext + ".txt"
    file_name_new = os.path.join(train_new_labels_dir, file_name_new)
    f = open(file_name_new, "w")
    f.write(new_data)
    f.close()

```
